```python
import pandas as pd
```

```python
df = pd.read_csv(r"찐최종원본.csv")
```

```python
df
```

```python
len(df.Channel_Title.unique())
```

```python
df['Comment_Published_At'] = pd.to_datetime(df['Comment_Published_At'])
df['Upload_Date'] = pd.to_datetime(df['Upload_Date'])
```

```python
df.info()
```

```python
# 필요없는 컬럼 제거
df = df.drop(['Comment_Count_5h','Comment_Count_10h','Comment_Count_24h','Time_Difference','Total_Replies','Comment_Like_Count','First_Video_Upload','Last_Video_Upload'],axis=1)
```

```python
df
```

```python
# 채널 주인의 댓글을 삭제
df = df[df['Is_Channel_Author'] != 1]
# 영상별로 동일한 작성자의 댓글을 1개만 남기기
df = df.drop_duplicates(subset=['Video_ID', 'Comment_Author'])
```

```python
# 댓글이 달린 시간과 동영상 업로드 시간의 차이 계산
df.loc[:, 'Time_Difference'] = (df['Comment_Published_At'] - df['Upload_Date']).dt.total_seconds() / 3600.0
```

```python
# 시간 차이가 일주 이내인 댓글 수 계산
df['Comment_Count_12h'] = df.groupby('Video_ID')['Time_Difference'].transform(lambda x: (x <= 168).sum())
```

```python
df
```

```python
df.reset_index(drop=True)
```

```python
df.info()
```

```python
df = df.drop(['Like_Count'],axis=1)
```

```python
# 'Channel_Title','Video_ID','Upload_Date','Comment_Published_At'순으로 정렬
df = df.sort_values(by=['Channel_Title','Video_ID','Upload_Date','Comment_Published_At'])
```

```python
df = df.reset_index(drop=True)
```
